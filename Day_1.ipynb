```python
while True:
    
    x = input('Enter the opertaion to be performed : ')
    t = np.linspace(0, np.pi, 100)
    if 'sin' in x.lower():
        y = np.sin(t)
        plt.plot(y)
        plt.show()
    elif 'cos' in x.lower():
        y = np.cos(t)
        plt.plot(y)
        plt.show()
    elif 'tan' in x.lower():
        y = np.tan(t)
        plt.plot(y)
        plt.show()
    elif 'log' in x.lower():
        y = np.log(t)
        plt.plot(y)
        plt.show()
    elif 'exp' in x.lower():
        y = np.exp(t)
        plt.plot(y)
        plt.show()
    elif -1 == int(x):
        break
    else:
        print("Enter sin,cos,tan,log,exp operation only")
```
